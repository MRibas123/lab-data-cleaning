## Global Historical Climatology Network Dataset
Variables are stored in both rows and columns
This dataset represents the daily weather records for a weather station (MX17004) in Mexico for five months in 2010.

```python
import pandas as pd
```

```python
weather = pd.read_csv('..\weather-raw.csv')
```

```python
weather.head()
```

```python
missing_values = weather.isnull().sum()
display(weather.shape, missing_values)
```

```python
weather = weather.drop(columns = ['d9','d12','d18','d19','d20','d21','d22','d24'])

weather.columns
```

```python
# weather = weather.fillna(0)



def get_temp(row):
    if row['element'] == 'tmax':
        return row[['d1', 'd2', 'd3', 'd4', 'd5', 'd6','d7', 'd8', 'd10', 'd11', 'd13', 'd14', 'd15', 'd16', 'd17', 'd23','d25', 'd26', 'd27', 'd28', 'd29', 'd30', 'd31']].max()
    elif row['element'] == 'tmin':
        return row[['d1', 'd2', 'd3', 'd4', 'd5', 'd6','d7', 'd8', 'd10', 'd11', 'd13', 'd14', 'd15', 'd16', 'd17', 'd23','d25', 'd26', 'd27', 'd28', 'd29', 'd30', 'd31']].min()
    
weather['d'] = weather.apply(get_temp,axis =1)

weather_fixed = weather[['id', 'year','month', 'element','d']]
```
